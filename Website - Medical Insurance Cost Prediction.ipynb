```python
import pickle
import warnings
warnings.filterwarnings('ignore')
from pywebio.input import *
from pywebio.output import *
from pywebio.platform.flask import webio_view
from flask import Flask
app = Flask(__name__)

def getStarted():
    with open('insurance_dataframe', 'rb') as f:
        df = pickle.load(f)
    with open('randomForestRegressor', 'rb') as f:
        randomForestRegressor = pickle.load(f)
    put_markdown('# Medical Insurance Cost Prediction')
    img = open('Medical Insurance.jpg', 'rb').read()
    put_image(img, width='50%')
    age = input('Input your age', type = NUMBER, required = True)
    sex = radio("What is your sex? ", options = ['Male', 'Female'], inline = True, required = True)
    if(sex == 'Male'):
        sex = 0
    else:
        sex = 1
    bmi = input('What is your bmi? ', type = FLOAT, required = True)
    children = input('How many children do you have? ', type = NUMBER, required = True)
    smoker = select("Are you a smoker? ", options = ['Yes', 'No'], required = True)
    if(smoker == 'Yes'):
        smoker = 1
    else:
        smoker = 0
    region = radio('Which region do you belong? ', options = ['southwest', 'southeast', 'northwest', 'northeast'], required = True)
    if(region == 'southwest'):
        region = 0
    elif(region == 'southeast'):
        region = 1
    elif(region == 'northwest'):
        region = 2
    else:
        region = 3
    
    predicted_charges = randomForestRegressor.predict([[age, sex, bmi, children, smoker, region]])
    
    put_markdown('## Predicted Medical Insurance Cost = `%s`' % predicted_charges[0])

app.add_url_rule('/insurance', 'webio_view', webio_view(getStarted), 
                 methods=['GET', 'POST', 'OPTIONS'])
app.run(host='localhost', port=80)
# http://localhost/insurance
```
